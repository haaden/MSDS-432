# 2. Calculate the factorial of each of the numbers using a simple FOR loop. Please capture the start/end times and store the results in an array, or linked list, or dictionary.

created a function that utilizes a for loop to calculate factorial of any number.
use a for loop to save the number, runtime and output (factorial) of this function for the 10 random numbers  in a dictionary

```python
# From the sample code provided and also show above
# Iterative  For Loop function to calculate factorial 
def fact_iter(x):
    ans = 1
    for i in range(1,x+1): 
        ans = ans * i 
    return ans
        
```

```python
# create a empty dictionary to record the runtime and factorial of for loop factorial function 
time_dict_itr = {}

# use for loop to iterate over the 10 numbers in the array
# calculate factorial for each number using the iterative factorial function 
# record runntime for each number 
for i in range(len(array_10)):
     i = int(i)
     start_time = time.perf_counter()
     it = fact_iter(array_10.item(i))
     end_time =  time.perf_counter() -start_time 
     time_dict_itr[array_10.item(i)]={"Runtime":end_time, "Factorial": it}
    

# output dict of for loop factorial fucntion runtime 
for k , v in time_dict_itr.items():
    print ("Number: ", k)
    print ("Runtime: ", v["Runtime"])
    print ("Factorial: ", v["Factorial"])
    print ("\n")
```

```python
# From the sample code provided and also show above
# Recursive function to calculate factorial 
def fact(x):
  if x == 1:
    return 1
  else:
    return x * fact(x-1)
```

```python

# create a empty dictinory to record the runtime and factorial of for loop factorial function 
time_dict_fact ={}

# use for loop to iterate over the 10 numbers in the array
# calculate factorial for each number using the recursive factorial function 
# record runntime for each number 
for i in range(len(array_10)):       
     start_time = time.perf_counter()
     f = fact(array_10.item(i))
     end_time = time.perf_counter() - start_time 
     time_dict_fact[array_10.item(i)]={"Runtime":end_time, "Factorial": f}
    

# output dict of recursive factorial fucntion runtime
for k , v in time_dict_fact.items():
    print ("Number: ", k)
    print ("Runtime: ", v["Runtime"])
    print ("Factorial: ", v["Factorial"])
    print ("\n")
```

# Discussion and Executive summary 


The graphic above shows that a iterative For Loop algorithm is much faster than a recursive algorithm for factorial calculations and the gap between the recursive algorithm and iterative For Loop algorithm widens as the number of the factorial being calculated becomes larger.

However, both algorithms have time complexity of O(n), which means that the algorithms runtime grows linearly. The difference between recursive and iterative For Loop functions is that iterative For Loop functions use a temp variable to store the product of the numbers that get updated in each loop. The recursive function saves the data for each function call in a stack until it reaches the end condition or the base case. After the end condition is met, the call stack in the memory is executed starting with the innermost call. 

This makes recursive function memory intensive, especially when calculating factorials as they produce large integers. It also seems that my  python version  recursive depth by default is set up to 3000 calls, which can be increased but it can easily run out of memory when dealing with extremely large numbers.

Also, the output of the factorial function can be excessively large and cause overflow issues in NumPy and Pandas objects as they use int64 or float64 (NumPy is mainly in C and some of Pandas is in C) which is not enough for very large numbers that we can get when calculating factorial for large numbers. I found two solutions to work around this problem: first, I saved the factorial in python int datatype where the value of an integer is not restricted by the number of bits and can expand to the limit of the available memory and second, I saved the number as strings in order to save it in Pandas dataframe. However, displaying the number (as string) in dataframe without truncation is not possible in the jupyter notebook. 

As data engineers, it is very important to know the limitations and benefits of a recursive algorithm and iterative algorithm. Iterative algorithms generally will be faster and recursive algorithms will be slower. Recursive algorithms should only be used if they do not significantly impact performance or cause errors due to memory limitation. However, recursive functions are simple and easy to read when compared to For Loops, which look messy. The read simplicity of recursive functions make them easier to maintain by programmers so the right approach should be evaluated on a case by case basis.
